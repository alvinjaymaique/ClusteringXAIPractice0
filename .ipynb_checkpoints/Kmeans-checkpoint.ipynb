```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

X, y, centers= make_blobs(n_samples=100, n_features=5, centers=3, random_state=42, cluster_std=4.0, return_centers=True)
```

```python
plt.scatter(X[:,2], X[:,3], c=y)
```

```python
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)
```

```python
plt.scatter(X_2d[:,0], X_2d[:,1], c=y)
plt.show()
```

```python
def clean_data(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*Q1
    upper_bound = Q3 + 1.5*Q3
    return data[(data >= lower_bound) & (data <= upper_bound)].reshape(-1,1)
```

```python
kmeans = KMeans(n_clusters=centers.shape[0], random_state=42) 
kmeans.fit(X) 
kmeans_distances = np.min(kmeans.transform(X), axis=1).reshape(-1,1)
cleaned_distances = clean_data(kmeans_distances)
percentile_90 = np.percentile(cleaned_distances, 90)
anomaly = (kmeans_distances > percentile_90)
```

```python
anomaly[(anomaly==True)]
```

```python
cmap = mcolors.ListedColormap(['blue', 'red'])
plt.scatter(X_2d[:,0], X_2d[:,1], c=anomaly, cmap=cmap)
# plt.colorbar()
plt.colorbar(label='(0: Normal, 1: Anomaly)')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, anomaly, test_size=0.2, random_state=42)
```

```python
# param_grid = {
#     'n_estimators': [100, 200, 500], 
#     'max_depth': [10, 20, None],    
#     'min_samples_split': [2, 10, 20],
#     'min_samples_leaf': [1, 5, 10],
#     'max_features': ['sqrt', 'log2'],
#     'class_weight': ['balanced', None]
# }

rfc_model = RandomForestClassifier(n_estimators=100, random_state=42)
rfc_model.fit(X_train, y_train.reshape(-1))
```

```python
y_pred = rfc_model.predict(X_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Predicted Negative', 'Predicted Positive',], 
            yticklabels=['True Negative', 'True Positive'])
plt.ylabel('True Anomaly')
plt.xlabel('Predicted Anomaly')
plt.title('Confusion Matrix')
plt.show()
```
